## Introduction

**Python** and **R** have been two popular programming languages for data scientists. However, data visualization in **Python** has long been refered to as painful and ugly compared to the handy tools available in **R**. This tutorial talks about how to use **ggplot** for visualization in **Python**. We will introduce common functions with examples that have been covered in previous lectures of 5702 but in **Python** this time.

## Histogram

```python
from plotnine import ggplot, aes, stat_bin, geom_histogram
huron.head(5)
```

```python
(ggplot(huron) + 
 geom_histogram(aes(x="level"), fill = 'cornflowerblue' ))
```

Notice that unlike in **R**, you need to use **quotes** when specifying the attribute name you want to plot.

## Boxplot

```python
from plotnine import geom_boxplot

(
  ggplot(huron)
  + aes(x="factor(decade)", y="level")
  + geom_boxplot()
)
```

## Density plots

```python
from plotnine import geom_density
mpg1 = pd.melt(mpg[['hwy', 'cty']])

ggplot(mpg1, aes(x='value', color='variable')) + \
    geom_density()
# ggplot(world, aes(x = GDP, y = ..density..))
```

## Violin plots

```python
from plotnine import geom_violin
ggplot(huron, aes(x="factor(decade)", y="level")) + \
    geom_violin() 
```

## Cleveland dot plot

```python
from plotnine import geom_point
from ggplot import *
(ggplot(economics, aes(y= 'date', x = 'unemploy')) +
  geom_point(color = "blue") + scale_y_reverse()
)
```

```python
mpg2 = mpg.groupby(['manufacturer', 'class']).mean().reset_index()
(ggplot(mpg2, aes(y= 'manufacturer', x = 'displ', 
                  color = 'class')) +
  geom_point()+scale_color_brewer(type='qual')
)
```

## Facet_wrap

```python
ggplot(mpg, aes(x='displ')) + \
    geom_histogram() + \
    facet_wrap("manufacturer")
```
